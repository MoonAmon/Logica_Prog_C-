```python
# algoritmo 174
def matrizBiOrdinaria(NumColuna, NumLinhas, elemento=0):
    matriz = [[0 for i in range(NumColuna)] for j in range(NumLinhas)]
    for y in range(NumColuna):
        for x in range(NumLinhas):
            matriz[x][y] = elemento
    return matriz
```

```python
# algoritmo 189
def convertCelsius(limitS, limitI, decremento):
    defir = limitS - limitI
    table2d = [[None for _ in range(2)] for _ in range(defir)]
    print(table2d)
    table2d[0][0] = 'Fahrenheit'
    table2d[0][1] = 'Celsius'
    for i in range(1,defir):
        for j in range(0,2):
            table2d[i][1] = limitS
            table2d[i][0] = round(limitS * (9/5) + 32, 2)
            limitS -= decremento
    return table2d
resultado_celsius = convertCelsius(100, 32, 1)
df_temps = pd.DataFrame(resultado_celsius)
df_temps.head(200)
```

```python
def create_matrix(colunms, rows):
    """Create a matrix filled with zeros.

    Args:
        colunms (int): The number of columns in the matrix.
        rows (int): The number of rows in the matrix.

    Returns:
        numpy.ndarray: A matrix with the specified number of columns and rows, filled with zeros."""
    return np.zeros((colunms, rows), dtype=object)

testfinal = create_matrix(10, 2)
testfinal = pd.DataFrame(testfinal)
testfinal.head(10)
```

```python
def data_people_matrix():
    """Fill a numpy matrix with input names, age and sex,
    and return data of male people with age 21 or more.

    Returns:
        numpy.ndarray: A numpy matrix with the input data ordered."""
    data = create_matrix(3,20)

    for x in range(20):
        data[x, 0] = input("Digite o nome: ")
        data[x, 1] = int(input("Digite a idade: "))
        data[x, 2] = str(input("Digite o sexo (M para masculino ou F para feminino): "))

    for x in range(20):
        if data[x, 1] > 21 and data[x, 2] == 'M':
            print("\nNome: {}".format(data[x, 0]))

data_people_matrix()
```

```python
# algoritmo 206
def find_higher_lower(array):
    """"""
    higher = array[0]
    lower = array[0]
    for i in array:
        if i > higher:
            higher = i
        if i < lower:
            lower = i
    
    return higher, lower

def get_array(count_array):
    """"""
    array = np.zeros(count_array)
    for i in range(count_array):
        array[i] = float(input("Digite o número: "))
    
    return array

def len_array():
    """"""
    count_array = int(input("Digite quantos números quer comparar: "))
    return count_array

def algoritmh_higher_lower():
    """"""
    count_array = len_array()
    array = get_array(count_array)
    higher, lower = find_higher_lower(array)
    print("Maior número digitado: {}  Menor número digitado: {}".format(higher, lower))

algoritmh_higher_lower()
```

```python
# algoritmo 209
def ricci_n(n_count):
    A = 1
    B = 2
    C = 3

    print("{}, \n{}, ".format(A, B))

    for _ in range(n_count):
        C = A + B
        print("{}, ".format(C))
        A = C - A
        B = C

def seque_ricci():
    n_count = int(input("Digite n número para n primeiros núemros da sequencia de racci: "))
    ricci_n(n_count)

seque_ricci()
```

```python
# algoritmo 212
def get_nums():
    array = np.zeros(20)
    for i in range(20):
        array[i] = float(input("Digite o número: "))
    return array

def sum_sqrt(array):
    sum = 0
    for i in array:
        if math.sqrt(i) < 225:
            sum += math.sqrt(i)

    return print("soma: {}".format(sum))

def algo_sum_sqrt():
    array = get_nums()
    sum_sqrt(array)

algo_sum_sqrt()
```

```python
# algoritmo 214
def get_grade_frame(lenght):
    grade_frame = create_matrix(3, lenght) 
    for i in range(lenght):
        grade_frame[0, i] = str(input("Digite o nome do aluno: "))
        grade_frame[1, i] = float(input("Digite a nota PR1: "))
        grade_frame[2, i] = float(input("Digite a nota PR2: "))
    
    grade_frame = pd.DataFrame(grade_frame.T, columns=["Nome", "PR1", "PR2"])
    grade_frame.head(lenght)

    return grade_frame

test = get_grade_frame(3)
test.head()
```

```python
# algoritmo 217
def get_names(length):
    names_frame = create_matrix(2, length)
    for i in range(length):
        names_frame[0, i] = str(input("Digite o nome: "))
        names_frame[1, i] = len(str(names_frame[0, i]))
    
    names_frame = pd.DataFrame(names_frame.T, columns=["Nome", "Qnt_letras"])
    return names_frame

test_count = get_names(8)
test_count.head(8)
```
