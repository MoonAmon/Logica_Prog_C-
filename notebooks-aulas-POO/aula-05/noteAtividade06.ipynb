```python
# algoritmo 349
def generate_arrays():
    array1 = [random.randint(1, 10) for _ in range(5)]
    array2 = [random.randint(1, 10) for _ in range(5)]
    return array1, array2

def generate_sum_array(array1, array2):
    sum_array = [array1[i] + array2[i] for i in range(5)]
    return sum_array

array1, array2 = generate_arrays()
sum_array = generate_sum_array(array1, array2)

print("Array 1:", array1)
print("Array 2:", array2)
print("Sum array:", sum_array)
```

```python
# algoritmo 351
def generate_names_array(length=5):
    names = [random.choice(['Alice', 'Bob', 'Charlie', 'David', 'Emily']) for _ in range(length)]
    return names

def find_name_position(name, names_array):
    try:
        position = names_array.index(name)
        return position
    except ValueError:
        return -1

names_array = generate_names_array()

print("Names array:", names_array)

number = int(input("Enter a number between 1 and 5: "))
if number < 1 or number > 5:
    print("Invalid number")
else:
    name = names_array[number-1]
    print(f"The name at position {number} is {name}")


# algoritmo 352
search_name = input("Enter a name to search for: ")
position = find_name_position(search_name, names_array)
if position == -1:
    print(f"{search_name} not found")
else:
    print(f"{search_name} found at position {position+1}")
```

```python
# algoritmo 354
def generate_numbers_array(length=15):
    numbers = [random.randint(1, 100) for _ in range(length)]
    return numbers

def get_parity(number):
    if number % 2 == 0:
        return "even"
    else:
        return "odd"

numbers_array = generate_numbers_array()

for i, number in enumerate(numbers_array):
    parity = get_parity(number)
    print(f"{i+1}. {number} ({parity})")
```

```python
# algoritmo 374
def generate_numbers_array(length=50):
    numbers = [random.randint(1, 100) for _ in range(length)]
    return numbers

def square_numbers(numbers_array):
    squared_array = [number**2 for number in numbers_array]
    return squared_array

numbers_array = generate_numbers_array()
squared_array = square_numbers(numbers_array)

square_dict = {
    'Number': numbers_array,
    'Square': squared_array
}

square_df = pd.DataFrame(square_dict)
square_df.head(50)
```

```python
# algoritmo 356
def generate_names_list(length=15):
    names = [random.choice(['Alice', 'Bob', 'Charlie', 'David', 'Emily', 'Frank', 'Grace', 'Henry', 'Isabella', 'Jack', 'Kate', 'Liam', 'Mia', 'Nathan', 'Olivia']) for _ in range(length)]
    return names

def generate_grades_list(length=15):
    grades = [[random.randint(0, 10), random.randint(0, 10)] for _ in range(length)]
    return grades

def calculate_average(grades):
    averages = [round(sum(grade)/2, 2) for grade in grades]
    return averages

def calculate_status(averages):
    status = ["AP" if average >= 6 else "RP" for average in averages]
    return status

def generate_students_dataframe(names, grades, averages, status):
    dict_students = {
        'Nome': names,
        'PR_1': [grade[0] for grade in grades],
        'PR_2': [grade[1] for grade in grades],
        'Media_final': averages,
        'Situação': status
    }
    return pd.DataFrame(dict_students)

names_list = generate_names_list()
grades_list = generate_grades_list()
averages_list = calculate_average(grades_list)
status_list = calculate_status(averages_list)
students_dataframe = generate_students_dataframe(names_list, grades_list, averages_list, status_list)

students_dataframe.head(15)
```

```python
# algoritmo 357
def generate_names_list(length=20):
    names = [random.choice(['Alice', 'Bob', 'Charlie', 'David', 'Emily', 'Frank', 'Mrace', 'Senry', 'Isabella', 'Lack', 'Kate', 'Liam', 'Mia', 'Nathan', 'Olivia', 'Paul', 'Quinn', 'Rachel', 'Sam', 'Tom']) for _ in range(length)]
    return names

def generate_salaries_list(length=20):
    salaries = [random.randint(1000, 5000) for _ in range(length)]
    return salaries

def calculate_new_salaries(salaries):
    new_salaries = [salary * 1.08 for salary in salaries]
    return new_salaries

def generate_employees_dataframe(names, salaries, new_salaries):
    dict_employees = {
        'Nome': names,
        'Salário': salaries,
        'Novo salário': new_salaries
    }
    return pd.DataFrame(dict_employees)

names_list = generate_names_list()
salaries_list = generate_salaries_list()
new_salaries_list = calculate_new_salaries(salaries_list)
employees_dataframe = generate_employees_dataframe(names_list, salaries_list, new_salaries_list)

employees_dataframe.head()
```

```python
# algoritmo 359
def generate_names_interval(names):
    names_interval = [name for name in names if name[0] >= 'L' and name[0] <= 'S']
    return names_interval

names = ['Loubert', 'Josuke', 'Mario', 'Chooper', 'Luffy', 'Buggy']
names_interval = generate_names_interval(names)
print(names_interval)
```
