```python
# algoritmo 379
def is_equal_30(vetor_numbers):
    count_30 = 0
    for number in vetor_numbers:
        if number == 30:
            count_30 += 1
    return count_30
   

def is_greater_mean(vetor_numbers):
    count_greater = 0
    mean_value = mean_vetor(vetor_numbers)
    for number in vetor_numbers:
        if number > mean_value:
            count_greater += 1
    return count_greater
   

def is_equal_mean(vetor_numbers):
    count_equal = 0
    mean_value = mean_vetor(vetor_numbers)
    for number in vetor_numbers:
        if number == mean_value:
            count_equal += 1
    return count_equal
 
def mean_vetor(vetor_numbers):
    vetor_mean = sum(vetor_numbers) // len(vetor_numbers)
    return vetor_mean
   

def algoritmo_number_comp():
    vetor_numbers = [random.randint(1,100) for _ in range(100)] 
    count_equal_30 = is_equal_30(vetor_numbers)
    count_greater = is_greater_mean(vetor_numbers)
    count_equal = is_equal_mean(vetor_numbers)
    
    print("{} número(s) lidos são igual(is) a 30 \n {} número(s) são maior que a média \n {} número(s) são igual(is) a média".format(count_equal_30, count_greater, count_equal))
    print(vetor_numbers)

algoritmo_number_comp()
```

```python
# algoritmo 389
def temp_month_array():
    temp_month = []
    for _ in range(12):
        temp_month.append(float(input("Digite a temperatura média do mês:")))
    return temp_month

def print_temp_month(temp_month):
    for index, temp in enumerate(temp_month):
        print("A temperatura média do mês {} é {}ºC".format(index + 1, temp))

def maior_menor_month_print(temp_month):
    maior = max(temp_month)
    menor = min(temp_month)
    index_maior = temp_month.index(maior)
    index_menor = temp_month.index(menor)
    print("A maior temperatura do ano foi {}ºC no mês {} e a menor foi {}ºC no mês {}".format(maior, index_maior + 1, menor, index_menor + 1))

def algoritmo_temp_month():
    temp_month = temp_month_array()
    print_temp_month(temp_month)
    maior_menor_month_print(temp_month)

algoritmo_temp_month()
```

```python
# algoritmo 394
def create_names_array(count=50):
    first_names = ['Jubileu', 'Mario', 'Peach', 'Jorge', 'Matheus', 'Jotaro', 'Josuke', 'Dio', 'Toodle' , 'João', 'John', 'Isabela', 'Laura']
    last_names = ['Souza', 'Joestar', 'Kennedy', 'de Andrade', 'Reis', 'Fagundes', 'Moura', 'Hendrix', 'Ford', 'Martinez', 'Oliveira']
    names = []
    for _ in range(count):
        first_name = random.choice(first_names)
        last_name = random.choice(last_names)
        name = "{} {}".format(first_name, last_name)
        names.append(name)
    return names


def create_grades_array(count=50):
    grades_array = [random.randrange(1,10) for _ in range(count)]
    return grades_array


def calculate_weighted_average(grades1_array, grades2_array):
    mean_grades = [(3 * grade1 + 7 * grade2) / 10 for grade1, grade2 in zip(grades1_array, grades2_array)]
    return mean_grades

def create_df_grades(names_array, grades1_array, grades2_array, mean_grades):
    grades_students = {
        'Nome': names_array,
        'Nota 1': grades1_array,
        'Nora 2': grades2_array,
        'Média': mean_grades 
    }
    return pd.DataFrame(grades_students)


def algoritmo_grades_mean():
    names_array = create_names_array()
    grades1_array = create_grades_array()
    grades2_array = create_grades_array()
    means_array = calculate_weighted_average(grades1_array, grades2_array)
    grades_df = create_df_grades(names_array, grades1_array, grades2_array, means_array)
    
    return grades_df

grades_df = algoritmo_grades_mean()
grades_df.head(50)
```
